```python
import pandas as pd
from datetime import datetime
```

```python
month_df = pd.read_csv('monthly_sales.csv')
```

```python
month_df.head()
```

```python
month_df.rename(columns={'Month, Year':'time'}, inplace=True)
```

```python
month_df.time = month_df.time.apply(lambda x: datetime.strptime(x,'%b-%y'))
month_df.index = month_df.time
month_df.drop(columns=['time'],inplace=True)
```
